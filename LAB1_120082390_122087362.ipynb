```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from matplotlib import animation
from IPython.display import HTML
```

```
# gerar pontos para formar uma reta
npoints = 100 #número de pontos para desenhar
x = np.zeros(npoints) #coordenada x da bola
y = np.zeros(npoints) #coordenada y da bola
v = np.array([2.,7.]) #direção inicial
# p = np.array([0,0])   CRIA NUMPY ARRAY DE INTEIROS, PROBLEMA!!!
p = np.array([0.,0.]) #ponto inicial
alpha = 0 #passos

for k in range(0,npoints):
  x[k] = alpha*v[0] + p[0]
  y[k] = alpha*v[1] + p[1]

  alpha = alpha + 0.1
```

```
plt.figure(figsize=(6,6))
plt.plot(x,y,'r.')
plt.xlim(-10,10)
plt.ylim(-10,10)
```

Animação abaixo!!

```
# primeiro, vamos apagar a antiga figure
plt.close()

# mesmas atribuicoes...
fig = plt.figure()
ax = plt.axes(xlim=(-11, 11), ylim=(-11, 11))
line, = ax.plot([], [], 'ro', lw=2)
#npoints = 7
x = np.zeros(npoints)
y = np.zeros(npoints)
v = np.array([-3.,7.])
p = np.array([0.,0.])
alpha = 0

```

```
# funcao init, inicia somente com background
def init():
    line.set_data([],[])
    return line,

alphax = alphay = 0
constante_x = 0.1
constante_y = 0.1
# funcao animate, chamada sequencialmente. Funciona da mesma forma que um for i in range(frames)
def animate(i):
    global x, y, alphax , alphay, constante_x, constante_y, v, p
    x[i] = alphax*v[0] + p[0]
    y[i] = alphay*v[1] + p[1]


    if x[i] >= 10 or x[i] <= -10:
      constante_x*= (-1)
    if y[i] >= 10 or y[i] <= -10:
      constante_y*= -1

    alphax = alphax + constante_x
    alphay = alphay + constante_y

    line.set_data(x[i], y[i])
    return line,

# chama o animator.  blit=True so redesenha as partes que de fato mudaram.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=npoints, interval=40, blit=True)




# Creditos e agradecimentos a Jake Vanderplas
# email: vanderplas@astro.washington.edu
# website: http://jakevdp.github.com
```

```
# funcao gera a animacao na pagina html corrente!
HTML(anim.to_jshtml())
```
